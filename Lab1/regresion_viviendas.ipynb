```python
# ===============================
# CONVERTIR VARIABLES CATEGORICAS
# ===============================

# Convertimos variables categóricas a variables numéricas
data = pd.get_dummies(data)

# Mostrar dataset transformado
data.head()
```

```python
# ===============================
# SEPARAR VARIABLES
# ===============================

# X -> características
# y -> variable objetivo (precio)

X = data.drop("price", axis=1)
y = data["price"]

# Convertir todas las columnas a valores numéricos
X = X.apply(pd.to_numeric)

# Convertir a arrays de numpy
X = X.values
y = y.values

print("Numero de ejemplos:", len(y))
print("Numero de caracteristicas:", X.shape[1])
```

```python
# ===============================
# NORMALIZAR CARACTERISTICAS
# ===============================

def normalizarCaracteristicas(X):

    # Convertimos a tipo float para evitar errores
    X = X.astype(float)

    # Media de cada característica
    mu = np.mean(X, axis=0)

    # Desviación estándar
    sigma = np.std(X, axis=0)

    # Evitar división por cero
    sigma[sigma == 0] = 1

    # Aplicar normalización
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma


# Aplicar normalización
X_norm, mu, sigma = normalizarCaracteristicas(X)

print("Normalizacion completada correctamente")
```

```python
# ===============================
# AGREGAR COLUMNA DE UNOS
# ===============================

m = len(y)

# Agregamos columna de 1s al inicio
X_ready = np.concatenate([np.ones((m,1)), X_norm], axis=1)

print("Nueva dimension de X:", X_ready.shape)
```

```python
# ===============================
# FUNCION DE COSTO
# ===============================

def calcularCosto(X, y, theta):

    m = len(y)

    # Predicciones
    h = np.dot(X, theta)

    # Cálculo del error
    J = (1/(2*m)) * np.sum((h - y)**2)

    return J
```

```python
# ===============================
# DESCENSO POR GRADIENTE
# ===============================

def gradienteDescendente(X, y, theta, alpha, iteraciones):

    m = len(y)

    historial_costo = []

    for i in range(iteraciones):

        # Predicción del modelo
        h = np.dot(X, theta)

        # Actualización de parámetros theta
        theta = theta - (alpha/m) * np.dot(X.T, (h-y))

        # Guardar costo
        historial_costo.append(calcularCosto(X,y,theta))

    return theta, historial_costo
```

```python
# ===============================
# ENTRENAMIENTO DEL MODELO
# ===============================

# Inicializamos theta
theta = np.zeros(X_ready.shape[1])

# Parámetros del algoritmo
alpha = 0.01
iteraciones = 500

# Entrenar modelo
theta, historial_costo = gradienteDescendente(X_ready, y, theta, alpha, iteraciones)

print("Modelo entrenado")
print("Primeros valores de theta:")
print(theta[:10])
```

```python
# ===============================
# GRAFICO DE COSTO
# ===============================

plt.plot(historial_costo)

plt.xlabel("Iteraciones")
plt.ylabel("Costo")

plt.title("Evolucion de la funcion de costo")

plt.show()
```

```python
# ===============================
# REGRESION POLINOMICA
# ===============================

# Crear características al cuadrado
X_poly = X_norm ** 2

# Agregar columna de unos
X_poly = np.concatenate([np.ones((m,1)), X_poly], axis=1)

theta_poly = np.zeros(X_poly.shape[1])

# Entrenar modelo
theta_poly, costo_poly = gradienteDescendente(X_poly, y, theta_poly, 0.01, 500)

print("Modelo polinomico entrenado")
```

```python
# ===============================
# ECUACION NORMAL
# ===============================

# Convertir X a tipo float para evitar errores de tipo object
X_float = X.astype(float)

# Agregar columna de unos
X_normal = np.concatenate([np.ones((m,1)), X_float], axis=1)

# Calcular theta usando la ecuación normal
theta_normal = np.linalg.inv(X_normal.T @ X_normal) @ X_normal.T @ y

print("Theta calculado con ecuacion normal")
print(theta_normal[:10])
```

```python
# ===============================
# PREDICCIONES
# ===============================

# Realizar predicciones
predicciones = np.dot(X_ready, theta)

print("Primeras 10 predicciones:")
print(predicciones[:10])
```
